```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("pavansubhasht/ibm-hr-analytics-attrition-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Assuming 'path' from the previous cell contains the directory where the dataset was downloaded
# Get the list of files in the directory
files = os.listdir(path)

# Find the CSV file (assuming it's the only CSV file in the directory)
csv_file = [f for f in files if f.endswith('.csv')][0]

# Construct the full path to the CSV file
file_path = os.path.join(path, csv_file)

# Read the CSV file
df = pd.read_csv(file_path)
df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def separate_numeric_boxplots(df):
  numeric_cols = df.select_dtypes(include='number').columns
  for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()

# Assuming your DataFrame is named 'df'
separate_numeric_boxplots(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# --- Load Data (assumed already loaded in df) ---

# --- Encode Target ---
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# --- Encode Categorical Features ---
cat_cols = df.select_dtypes(include='object').columns
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

# --- Check for NaNs ---
assert df.isnull().sum().sum() == 0, "Dataset contains NaN values!"

# --- Feature/Label Split ---
X = df.drop('Attrition', axis=1)
y = df['Attrition']

# --- Scale Features ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- Train/Test Split ---
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# --- Build FNN Model ---
model = Sequential([
    Dense(64, activation='relu', input_dim=X_train.shape[1]),
    Dropout(0.4),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

# --- Compile Model ---
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# --- Define EarlyStopping ---
early_stopping = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)

# --- Train Model ---
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)

# --- Evaluate Model ---
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy:.4f}')
```

```python
plt.figure(figsize = (12,10))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accurcy')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(['train','Accuracy'])
plt.show()
```

```python
plt.figure(figsize = (12,10))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel("Epochs")
plt.ylabel("loss")
plt.legend(['train','loss'])
plt.show()
```
